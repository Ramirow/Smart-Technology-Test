```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt


df = pd.read_csv("task_data.csv", encoding='cp1252')


sh = df.corr()["class_label"]
sh = sh.reindex(sh.abs().sort_values(ascending=False).index)[1:]
print(sh)






```

```python
import seaborn as sns
colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
# plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(df.corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
plt.savefig('./correlation.png', dpi=None, facecolor='w',
                edgecolor='w',
                orientation='portrait', papertype=None, format=None,
                transparent=False, bbox_inches='tight', pad_inches=0.1,
                frameon=None, metadata=None)
```

```python
df = pd.read_csv("task_data.csv", encoding='cp1252')
y = df['class_label']
del df['class_label']
del df['sample index']

i=0
for col,v in sh.items():
    i+=1
    if i < 5:
        continue
    del df[col]
    
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score


model = LogisticRegression()

#cross validation
print('\n----\nCross validation\n')
scores = cross_val_score(model, df.astype(float), y, cv=5)
print(scores)
print("Accuracy: %0.2f " % scores.mean())
```
